```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
sns.pairplot(data, hue='Gender', palette='dark')
```

```python
dados_selecionados = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(random_state=0)
```

```python
kmeans.fit(dados_selecionados)
```

```python
kmeans.labels_
```

```python
fig, ax = plt.subplots()

ax.scatter(dados_selecionados['Annual Income (k$)'], dados_selecionados['Spending Score (1-100)'],c=kmeans.labels_)

plt.show()
```

```python
cotovelo = [] # SSE - SUM OF SQUARED ERRORS
 
k_values = range(1, 50)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(dados_selecionados)
    cotovelo.append(kmeans.inertia_)  # inertia_ da o mais perto ao centro "SSE decreases as the number of clusters increases. The "elbow" point indicates the optimal number of clusters"

fig, axis = plt.subplots()
axis.plot(k_values,cotovelo,marker='o')
plt.show()
```

```python
cotovelo = [] # SSE - SUM OF SQUARED ERRORS
 
k_values = range(1, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(dados_selecionados)
    cotovelo.append(kmeans.inertia_)  # inertia_ da o mais perto ao centro "SSE decreases as the number of clusters increases. The "elbow" point indicates the optimal number of clusters"

fig, axis = plt.subplots()
axis.plot(k_values,cotovelo,marker='o')
plt.show()
```

```python
kmeans3 = KMeans(n_clusters=5, random_state=0)
kmeans3.fit(dados_selecionados)
fig, ax = plt.subplots()

plt.scatter(dados_selecionados['Annual Income (k$)'], 
            dados_selecionados['Spending Score (1-100)'], 
            c=kmeans3.labels_, cmap='viridis', marker='o', alpha=0.5)

plt.show()
```

```python
# Retrieve the centroids
centroids = kmeans3.cluster_centers_

# Print the centroids
print("Centroids of the clusters:")
print(centroids)
```

```python
centroids = kmeans3.cluster_centers_

plt.figure(figsize=(8, 6))

plt.scatter(dados_selecionados['Annual Income (k$)'], 
            dados_selecionados['Spending Score (1-100)'], 
            c=kmeans3.labels_, cmap='viridis', marker='o', alpha=0.5)

plt.scatter(centroids[:, 0], centroids[:, 1], 
            c='red', s=300, alpha=0.75, marker='p', label='Centroids')

plt.title('Centroides')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
